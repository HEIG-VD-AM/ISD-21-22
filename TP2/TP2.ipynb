```python
import numpy as np

def isprime(num):
    for x in range(2, num):
        if num % x == 0 and num != 2 or num % np.sqrt(num) == 0:
            return False
    return True

array = np.arange(1,100)
for x in array + 1: 
    if isprime(x):
        print(x)
```

```python
import numpy as np
import matplotlib.pyplot as plt 

x = np.linspace(-5.0, 5.0)
y = (1 - np.exp(-2 * x)) / (1 + np.exp(-2 * x))
plt.plot(x, y)
plt.grid()
plt.title("tanh() function")
plt.xlabel("Values of x")
plt.ylabel("Values of y")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 

x = np.linspace(-5.0, 5.0)
y = (1 - np.exp(-2 * x)) / (1 + np.exp(-2 * x))
z = 1/(1 + np.exp(-x))

plt.figure()

#upper part of subplot 
plt.subplot(2, 1, 1)
plt.plot(x, y)
plt.title('tanh(x) function')
plt.xlabel('Values of x')
plt.ylabel('Values of y')
plt.grid()

#lower part of subplot
plt.subplot(2, 1, 2)
plt.plot(x, z)
plt.title('sigm(x) function')
plt.xlabel('Values of x')
plt.ylabel('Values of y')
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 
R1 = 1.1
w1 = 1.1
R2 = 0.8
w2 = 1.7
fi = 1.2
kfi = 0.008
k2 = -0.012
t= np.arange(-3.5, 35, 0.01)

x = R1*np.cos(2*np.pi*w1*t) + R2*np.sin(2*np.pi*w2*t+fi*np.exp(kfi*t))*np.exp(k2*t)
y = R1*np.sin(2*np.pi*w1*t) + R2*np.cos(2*np.pi*w2*t+fi*np.exp(kfi*t))*np.exp(k2*t)

plt.plot(x, y)
plt.grid()
plt.title("Parametric curve")
plt.xlabel("Values of x")
plt.ylabel("Values of y")
plt.figure(figsize=(12,12)) 
plt.show()

```

```python
import numpy as np
from matplotlib import pyplot as plt

def histo(array):
    array.sort()
    #print(array)
    
    frequency = [np.min(array)]
    binSize = (np.max(array) - np.min(array))/10
    for x in range(1, 10):
        frequency.append(binSize*x)  
    frequency.append(np.max(array))
    frequency = np.ceil(frequency)
    #print(frequency)
    
    counter = []
    for x in range(0, 10):
        count = 0
        if(x < 9):
            count = ((array >= frequency[x]) & (array < frequency[x+1])).sum()
        else:
            count = ((array >= frequency[x]) & (array <= frequency[x+1])).sum()
        counter.append(count)
    #print(counter)
    
    finalHisto = [frequency, counter]
    return finalHisto

a = np.random.randint(0,200,100) 
finalHisto = histo(a)    

tags = []
for i in range(len(finalHisto[0]) - 1):
    tags.append(str(int(finalHisto[0][i])) + "-" + str(int(finalHisto[0][i+1])))            
#print(tags)

y_pos = np.arange(len(tags))

plt.bar(y_pos, finalHisto[1], align='center', alpha=0.5)
plt.xticks(y_pos, tags)
plt.ylabel('Values')
plt.title('Handmade histogram')
plt.xticks(rotation=90)
plt.show()

plt.hist(a, bins = finalHisto[0]) 
plt.title("Python histogram") 
plt.show()
```
